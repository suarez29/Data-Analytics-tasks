# >El Ejercicio 1, es genere ventas de los ultimos dias y muestre por pantalla una serie(dataframe) con los datos de las ventas por productos.




```python
# importamos las librer√≠as
import numpy as np
import pandas as pd
```

```python
"""
Ejercicio 1 : Generar ventas de los ultimos 5 dias

1. creamos importamos las librerias numpy y pandas.
el panda utilisamos para crear un dataframe con los datos de las ventas y productos.
el numpy importamos para las calculos matematicos.

2.  creamos una dataframe
-creamos una semilla np.random.seed() utilisamos 1234510 como parametro. Pero puede ser otro numero
-llamamos a pandas y llamamos a dataframe con el nombre df1
-para el index utilisamos fecha con rango asta 10. esto qiere decir que la fecha comienza de la dia 1 asta 10.
-creamos una lista con elementos Ventas, Productos Unidades y precio a unidad
- a el df1 (dataframe) utilizando assign para asignar a un lista de productos a la columna productos
- a el df1 (dataframe) utilizando assign para asignar a un lista de unidades a la columna unidades
- a el df1 (dataframe) utilizando assign para asignar a un lista de precio a unidad a la columna precio_unidad
- para calcular utilizamos a labda y asignamos el calculo de Unidades * Precio_unidad a la columna Ventas
"""


np.random.seed(1234510)

df1 = pd.DataFrame(
        index=pd.date_range("20220101", periods=10),
        columns = ['Ventas','Productos', 'Unidades', 'Precio_unidad',]
        )
df1 = df1.assign(Productos=['Ventilador','Direct Tv','Cama','Labadora','bycicleta','armario','sartenes','sofa','iphone','estufa'])
df1 = df1.assign(Unidades=[50,60,20,20,30,20,10,45,100,10])
df1 = df1.assign(Precio_unidad=[100,1200,2000,3000,1000,3000,700,60,1000,900])

df1=(
    df1
    .assign(Ventas= lambda x: x.Unidades*x.Precio_unidad)

)

df1
```

```python
df1[5:] # Para llamar los ultimos ventas llamamos el dataframe con el nombre df1, y despues con el index 5: a 10 para las 5 ultimos dias
```

# Ejercicio 2
Reciba un dictionario con los ingresos mensuales de los clientes de una impresa, y devuelva una serie con ingreso minimo, maximo y media

```python
"""""
Ejercicio 2
Aqui tambien importamos las librerias numpy y pandas para los calculos matematicos,
y pandas para crear el dataframe.

1. creamos un diccionario con los ingresos mensuales de los clientes de una impresa
En un dictionario se tiene un key y un valor, el key es el funcion del cliente y el valor es el ingreso mensual.
2. creamos contenedores para los calculos matematicos
Para la suma total utitlizamos sum ,con el nombre del dictionario y sus valores para calcular el total de ingresos
3. calculamos la media con utilizando el calculo total de ingresos y devidido entre (len) que el cantidad de los keys
para el minimo utilizamos min con el parametro del dictionario y sus valores
para el maximo utilizamos max con el parametro del dictionario y sus valores
"""

import numpy as np
import pandas as pd

dicIngreso = {
    "Director": 16000,
    "Secretaria":5000,
    "Vendedor": 10000,
    "Mecanico": 8000,
    "Vendedor de priesas": 12000,
}


total = sum(dicIngreso.values()) #creamos una variable para el calculo de la suma total
media = total/len(dicIngreso) #creamos una variable para el calculo de la media
minimo = min(dicIngreso.values())#creamos una variable para el calculo minimal
maximo = max(dicIngreso.values())# creamos una variable para el calculo max
a = (total,media,minimo,maximo) #creamos una tupla con los calculos matematicos
pd.Series(a)# el pd de pandas para crear los datos en una series
```

# Ejercicio 3
Ecribir programa que genere y muestre por pantalla un Dataframe con los datos

```python
"""
Ejercicio 3
Aqui tambien importamos las librerias numpy y pandas para los calculos matematicos,
y pandas para crear el dataframe.

1. creamos un dictionario con una lista de datos
2. creamos un dataframe con el nombre df2
"""
import numpy as np
import pandas as pd
Empresa = {
    "Mes": ["Enero","Febrero","Marzo","Abril","Mayo"],
    "Ventas":[30500, 35600, 28300, 33900, 42500],
    "Gastos" : [22000, 23450, 18100, 35700, 32450]
}

df2 = pd.DataFrame(Empresa)


df2
```

# Ejercicio 3.1
Columna 1 = Gasto/Ventas*100


```python
df2['Columna1'] = df2['Gastos'] / df2['Ventas']*100 # agregamos en la lista  de las columna el nombre Columna1 y con el nombre del las columnas Gastos y Ventas dividimos y despues muliplicamos
df2
```

#Ejercicio 3.2
Columna 2 = Si Ventas>= 30000, imprimir Meta superada, caso contrario imprimir Meta no superada

```python
"""
Como anterior nececitamos a la libreria numpy para los calculos matematicos
1. creamos un funcion de assign los valores de la columna Ventas
aque utilizamos if else statement para asignar los valores a la columna Columna2
si Ventas >= 30000, imprimir Meta superada, caso contrario imprimir Meta no superada
"""
import numpy as np
def assign_value(Ventas):
    if Ventas >= 30000:
        return "Meta Superada"
    else:
        return "Meta No Superada"
"""
con el apply aplicamos la funcion a la columna Ventas para que asignar los valores a la columna Columna2
"""
df2['Columna2'] = df2['Ventas'].apply(assign_value)
df2
```

#Ejercicio 3.3
Calcula el sobre gasto de los ultimos 3 meses


```python
"""

1. Utilisamos assign para asignar a la columna(columna3) sobre gasto el calculo de Gastos - Ventas
- Con Columna 3 crea una columna en el dataframe para guardar el calculo de Gastos - Ventas
"""
import numpy as np
df2=(
    df2
    .assign(Columna3 = lambda x: x.Gastos - x.Ventas)
)
df2



```

# Ejercicio 3.4
Obtener el total de las ventas de los ultimos 3 meses

```python
"""
con el nombre tail guaramos en la variable tail las ultimas 3 filas del dataframe df2
despues con el nombre ultimos_ventas utilizamos el metodo sum para sumar los valores de la columna Ventas
"""
tail = df2.tail(3)# el tail muestra por default las ultimas 5 filas, si nostros ponemos 3, le saca los ultimos 3
ultimo_ventas = tail['Ventas'].sum()# para el total los ultimos 3 meses de las ventas utilizamos .sum para sumar los elementos dentro del columna ventas
print('Total ventas de los ultimos 3 meses es', ultimo_ventas)# para informar utilizamos print con un string y el nombre del contenedor
```
